#프로그램 설계 tip

프로그램을 간단하게 나누면 

1. 입력
2. 처리
3. 출력

으로 나눌 수 있다. 

이렇게 부분을 나누고 먼저 생각하면 프로그램 설계가 편해진다.

예를 들어 cm를 인치로 변환하는 프로그램을 생각해보자.

```python
#입력
a = input("cm를 입력하세요 ")
a = float(a)
#처리
a = a * 0.393701

#출력
print( a,"inch입니다")
```

#split function

하나의 문자열을 여러개로 짤라서 표현이 가능하다.

list 안에 원하는 문자열들이 담겨서 출력 됨

```python
print("10 20 30 40".split(" "))
print("10-20-30-40".split("-"))
#기본적으로 ()에 아무 값을 넣지 않으면 공백으로 처리함.
print("10   20\t  30\n   40".split()) 
```
